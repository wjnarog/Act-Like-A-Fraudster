```python
import requests
from bs4 import BeautifulSoup

#First, scrape the webpage which is done here with requests python lib
URL = "https://realpython.github.io/fake-jobs/"
page = requests.get(URL)

#Next, parse the results with a tool like beautiful soup
soup = BeautifulSoup(page.content, "html.parser")
```

```python
#Find Elements by ID
results=soup.find(id="ResultsContainer")
```

```python
print(results.prettify)
```

```python
#Find Elements by Class name
job_cards = results.find_all("div", class_="card-content")
```

```python
print(job_cards)
```

```python
#Pick out elements with Find
for job_card in job_cards:
    title_element = job_card.find("h2", class_="title")
    company_element = job_card.find("h3", class_="company")
    location_element = job_card.find("p", class_="location")
    print(title_element)
    print(company_element)
    print(location_element)
    print()
```

```python
#Further Extract Text
for job_card in job_cards:
    title_element = job_card.find("h2", class_="title")
    company_element = job_card.find("h3", class_="company")
    location_element = job_card.find("p", class_="location")
    print(title_element.text.strip())
    print(company_element.text.strip())
    print(location_element.text.strip())
    print()
```

```python
URL = "https://www.python.org/jobs/?page=1"
page=requests.get(URL)
list = []

soup1 = BeautifulSoup(page.content, "html.parser")
results = soup1.find(id="content")

results = results.find(class_="list-recent-jobs list-row-container menu")
job_cards= results.find_all("li")
for job_card in job_cards:
    job_title = job_card.find(class_="listing-company-name")
    if job_title is None:
        continue
    job_title = job_title.find('a')
    list+=job_title
print(list)
```
