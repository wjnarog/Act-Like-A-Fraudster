```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Firefox()
driver.get("https://www.landwatch.com/colorado-land-for-sale/undeveloped-land")
#print(driver.page_source)
soup2 = BeautifulSoup(driver.page_source, "html.parser")
driver.quit()
results=soup2.find(class_="_8f63d fc97d")
listings=results.find_all(id="placard-container")

#for listing in listings:
#    listing = listing.find()
```

```python
for listing in listings:
    #listingGlanceInfo = listing.find('script')
    listingPrice = listing.find(class_="_6ae86")
    listingAcreage = listing.find(class_="_6a6db")
    listingAddress = listing.find("p")
    listingAgent = listing.find(id="placard-broker").find(class_="daf63").find(class_="_541a6")
    listingCompany = listing.find(id="placard-broker").find(class_="daf63").find(class_="_61da4")
    
    if listingPrice:
        print(listingPrice.text.strip())
    else:
        print("None")
        
    if listingAcreage:
        print(listingAcreage.text.strip())
    else:
        print("None")

    if listingAddress:
        print(listingAddress.text.strip())
    else:
        print("None")

    if listingAgent:
        print(listingAgent.text.strip())
    else:
        print("None")

    if listingCompany:
        print(listingCompany.text.strip())
    else:
        print("None")
    print("")
    
```
